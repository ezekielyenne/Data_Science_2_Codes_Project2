**Importing Modules for the Task**

```python
## Importing all the needed modules
# Import pandas and numpy
import pandas as pd
import numpy as np
# Import seaborn
import seaborn as sns
sns.set_theme()
# Import matplotlib
import matplotlib.pyplot as plt
import datetime
```

**Data Extraction, Transformation, Loading and Analyses**

```python
# Load the data
HR_data = pd.read_csv('/HRDataset_v14.csv')
HR_data
```

```python
# Checking the column information
HR_data.info()
```

```python
## check the columns
HR_data.columns
```

```python
## Codes to minimize memory size and assign the right datatypes to fields

number_columns = ['EmpID', 'MarriedID', 'MaritalStatusID', 'GenderID',
       'EmpStatusID', 'DeptID', 'PerfScoreID', 'Zip', 'FromDiversityJobFairID',
       'Salary', 'Termd', 'PositionID', 'EmpSatisfaction', 'SpecialProjectsCount',
       'DaysLateLast30', 'Absences']
for column in number_columns:
    HR_data[column] = HR_data[column].astype("int32")

number_columns = ['Employee_Name','Position', 'State', 'DOB',
       'Sex', 'MaritalDesc', 'CitizenDesc', 'HispanicLatino', 'DateofHire',
       'DateofHire.1', 'TermReason', 'EmploymentStatus', 'Department',
       'ManagerName', 'RecruitmentSource', 'PerformanceScore', 'LastPerformanceReview_Date']
for column in number_columns:
    HR_data[column] = HR_data[column].astype("string")

number_columns = ['ManagerID','EngagementSurvey']
for column in number_columns:
    HR_data[column] = HR_data[column].astype("float32")

HR_data.info()
```

```python
## Checking the shape of the data
HR_data.shape
```

```python
## Descriptive statistics of relevant numeric fields
HR_data.describe()
```

```python
# Count columns to know missing values
HR_data.count()
```

```python
# Forward-fill missing values (use the previous non-missing value):
HR_data_filled=HR_data.fillna(method='ffill',axis=1)
```

```python
# Columns count
HR_data_filled.count()
```

```python
#Histogram of the major continuous column - Most worker's salary fall within 50, 000 to 100, 000
fig = plt.figure(figsize=(8,5))
g = sns.distplot(HR_data.Salary)
g.set_title('Salary Distribution');
```

**Data** **Visualization** **and** **Interpretation**

```python
## Check the unique recruitment sources for the company
HR_data_filled['RecruitmentSource'].unique()
```

```python
# Number of staff per recruiting agency
fig = plt.figure(figsize=(8,5))
g = sns.catplot(x="RecruitmentSource",kind='count',order=['LinkedIn', 'Indeed', 'Google Search',
       'Employee Referral', 'Diversity Job Fair', 'On-line Web application', 'CareerBuilder',
       'Website', 'Other'], aspect=3.3, data=HR_data_filled);
g.fig.suptitle("Count of Performance of Agency Recruitment",y=1.05);
```

```python
# First four best performing recruiting companies
fig = plt.figure(figsize=(8,5))
g = sns.catplot(x="RecruitmentSource",kind='count',order=['LinkedIn', 'Indeed', 'Google Search','Employee Referral' ], aspect=1.5, data=HR_data_filled);
g.fig.suptitle("Four Best Performing Recruiting Agencies",y=1.05);
```

```python
# Checking the unique number of Marital status in the company
HR_data_filled['MaritalDesc'].unique()
```

```python
# Checking marital status of the company
fig = plt.figure(figsize=(8,5))
g = sns.catplot(x="MaritalDesc",kind='count',order=['Single', 'Married', 'Divorced', 'Widowed', 'Separated'], aspect=1.3, data=HR_data_filled);
g.fig.suptitle("Count of Marital status",y=1.05);
```

```python
## Checking the number of Males and Females
HR_data_filled['Sex'].unique()
```

```python
# Number of staff by gender
fig = plt.figure(figsize=(8,5))
g = sns.catplot(x="Sex",kind='count',order=['M ', 'F'], aspect=1.3, data=HR_data_filled);
g.fig.suptitle("Count for Gender",y=1.05);
```

```python
## Check unique departments
HR_data_filled['Department'].unique()
```

```python
## Group by department
HR_data_filled_grouped=HR_data_filled.groupby(['Department']).sum()
```

```python
# Checking departmental Performance i.e, best performing (Production department best performing)
plt.barh(HR_data_filled_grouped.index,HR_data_filled_grouped['PerfScoreID'])
```

```python
# Get the department with the highest salaries pay
plt.barh(HR_data_filled_grouped.index,HR_data_filled_grouped['Salary'])
```

```python
# Number of Staff per department
fig = plt.figure(figsize=(10,4))
g = sns.catplot(x="Department",kind='count',order=['Production       ', 'IT/IS', 'Software Engineering',
       'Admin Offices', 'Sales', 'Executive Office'], aspect=2.0, data=HR_data_filled);
g.fig.suptitle("Number of staff per department",y=1.05);
```

```python
# Checking which sex group has the best performance
HR_data_filled_grouped_sex=HR_data_filled.groupby(['Sex']).sum()
```

```python
# Performance within the department per sex
plt.bar(HR_data_filled_grouped_sex.index,HR_data_filled_grouped_sex['PerfScoreID'])
```

```python
# Converting the performance review date to datetime datatype
HR_data_filled['LastPerformanceReview_Date'] = pd.to_datetime(HR_data_filled['LastPerformanceReview_Date'])
HR_data_filled.head(5)
```

```python
 #line plot to show the trend in prices over time
fig = plt.figure(figsize=(8,3))
g = sns.lineplot(x="LastPerformanceReview_Date_new",y="Absences",data=HR_data_filled)
g.set_title('Employees performance based on Absences')
```

```python
# Clean and manipulate the Date of Hire
HR_data_filled_Hire = HR_data_filled['DateofHire']

# Convert the column to datetime type
HR_data_filled['DateofHire_Clean'] = pd.to_datetime(HR_data_filled['DateofHire'])

# Extract only the year
HR_data_filled['DateofHire_Clean_Year'] = HR_data_filled['DateofHire_Clean'].dt.year
HR_data_filled.head(5)
```

```python
# Unique years for date of hire
HR_data_filled['DateofHire_Clean_Year'].unique()
```

```python
 #line plot to show the trend in prices over time
fig = plt.figure(figsize=(8,3))
g = sns.lineplot(x="DateofHire_Clean_Year",y="Salary",data=HR_data_filled)
g.set_title('Trend in Salary Allocation')
```

```python
# Number of staff by hire date
fig = plt.figure(figsize=(8,5))
g = sns.catplot(x="DateofHire_Clean_Year",kind='count',order=[2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018], aspect=2.3, data=HR_data_filled);
g.fig.suptitle("Count of Number of hire per year",y=1.05);
```

```python
## Mean Salary per department
plt.figure(figsize=(12, 6))
sns.barplot(x='Department', y='Salary', width = 0.5, data=HR_data_filled)
plt.title('Grouped Data Visualization')
plt.xlabel('Department')
plt.ylabel('Mean Salary per Department')
plt.show()
```
